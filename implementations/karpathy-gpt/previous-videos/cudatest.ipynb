```python
import torch

import matplotlib.pyplot as plt
```

```python
# torch gausian random tensor of length 100
x = torch.randn(2000)

hy, hx = torch.histogram(x, bins=100)

# plot histogram
plt.plot(hx[:-1], hy)

iterations = 5
x.view(2000, 1) @ torch.ones(1, iterations)
```

```python
samples = 1000
a = torch.randn(samples, 1, 10)
b = torch.randn(samples, 10, 1)

mul = (a @ b).view(samples)


# Assuming you have multiple tensors for which you want to create histograms
tensors = [a, mul]

plt.figure(figsize=(20, 4)) # width and height of the plot
legends = []
for i, tensor in enumerate(tensors):
    t = tensor.view(-1).to('cpu') # flatten the tensor and move it to cpu
    print('tensor %d: mean %+.2f, std %.2f' % (i, t.mean(), t.std()))
    hy, hx = torch.histogram(t, bins=100)
    plt.plot(hx[:-1], hy)
    legends.append(f'tensor {i}')
plt.legend(legends)
plt.title('tensor distribution')
```
